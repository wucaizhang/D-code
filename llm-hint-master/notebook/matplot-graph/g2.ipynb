```python
import numpy as np
import matplotlib.pyplot as plt
 

figura = plt.figure(figsize=(7,7))
ejes = plt.subplot(111, projection="3d")

def f(x,y):
    return ((x**4)+(y**4)) 

x = np.linspace(-2,2,40) 
y = np.linspace(-2,2,40)

x,y = np.meshgrid(x,y)

z= f(x,y) 

ejes.plot_wireframe(x,y,z)
plt.show()
```
