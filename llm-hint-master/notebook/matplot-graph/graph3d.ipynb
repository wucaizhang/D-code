```python
import numpy as np
import matplotlib.pyplot as plt
 

figura = plt.figure(figsize=(7,7))
#               row col index
ejes = plt.subplot(1,1,1, projection="3d")


def f(x1,x2):
    return 5*x1 + 2*x2 + 2

x1 = np.linspace(1,6)
x2 = np.linspace(-1,5)
x3 = np.linspace(-9,-3)

x1,x2 = np.meshgrid(x1,x2)

y = f(x1,x2) 

ejes.plot_wireframe(x1,x2,y)
plt.show()
```
